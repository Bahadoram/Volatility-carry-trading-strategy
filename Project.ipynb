```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
import scipy as sp
import seaborn as sns
import pandas as pd
```

```python
file_name = 'grid1_zheb51fo.xlsx'
UX1       = pd.read_excel(file_name, sheet_name='UX1_Index')
UX2       = pd.read_excel(file_name, sheet_name='UX2_Index')
UX1
```

```python
fig,ax = plt.subplots(ncols=2, figsize=(15,7))
ax[0].plot(UX1.Date, UX1.PX_LAST)
ax[1].plot(UX2.Date, UX2.PX_LAST)
ax[0].set_title('UX1')
ax[1].set_title('UX2')
ax[0].grid(linestyle='dotted')
ax[1].grid(linestyle='dotted')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,7))
ax.plot(UX1.Date, UX1.PX_LAST, label='UX1', color='blue')
ax.plot(UX2.Date, UX2.PX_LAST, label='UX2', color='red')
ax.legend(loc='upper center', ncol=2)
ax.grid(linestyle='dotted')
plt.show()
```
